***Data_Cleaning***

```python
data.replace('?', pd.NA, inplace=True)
```

```python
data.isnull().sum()
```

```python
data['price'] = pd.to_numeric(data['price'], errors='coerce')
data['horsepower'] = pd.to_numeric(data['horsepower'], errors='coerce')
data['city-mpg'] = pd.to_numeric(data['city-mpg'], errors='coerce')
data['highway-mpg'] = pd.to_numeric(data['highway-mpg'], errors='coerce')
data['peak-rpm'] = pd.to_numeric(data['peak-rpm'],errors='coerce')
data['curb-weight'].fillna(data['curb-weight'].mean(), inplace=True)
data['make'].fillna('Unknown', inplace=True)
data['drive-wheels'].fillna('Unknown', inplace=True)
doors_mapping = {'two': 2, 'four': 4}
data['num-of-doors'] = data['num-of-doors'].map(doors_mapping)
```

Distribution_of_car_price

```python
car_prices = data['price']
less_than_10 = sum(1 for price in car_prices if price < 10000)
less_than_20 = sum(1 for price in car_prices if price < 20000 and price>10000)
less_than_30 = sum(1 for price in car_prices if price < 30000 and price>20000)
greater_than_30 = sum(1 for price in car_prices if price > 30000)

counts = [less_than_10, less_than_20, less_than_30, greater_than_30]
categories = ['< 10 lac', '< 20 lac', '< 30 lac',' 30 lac <']
plt.bar(categories, counts, color=['blue', 'green', 'red','purple'], alpha=0.7)

plt.xlabel('Price Range (in lacs)')
plt.ylabel('Number of Cars')
plt.title('Distribution of Car Prices')
for i, count in enumerate(counts):
    plt.text(i, count + 0.1, f'{count} cars', ha='center', va='bottom')


```

```python

car_prices = data['price']
less_than_10 = sum(1 for price in car_prices if price < 10000)
less_than_20 = sum(1 for price in car_prices if 10000 <= price < 20000)
less_than_30 = sum(1 for price in car_prices if 20000 <= price < 30000)
greater_than_30 = sum(1 for price in car_prices if price >= 30000)

cumulative_counts = [less_than_10,  less_than_20,  less_than_30, greater_than_30]
categories = ['< 10 lac', '< 20 lac', '< 30 lac',' 30 lac <']

plt.figure(figsize=(10, 6))
plt.plot(categories, cumulative_counts, marker='o', linestyle='-', color='blue')
plt.xlabel('Price Range (in lacs)')
plt.ylabel('Cumulative Number of Cars')
plt.title('Cumulative Distribution of Car Prices')

# Annotate the points with the cumulative counts
for i, count in enumerate(cumulative_counts):
    plt.text(i, count + 0.1, f'{count} cars', ha='center', va='bottom')

plt.show()
```

**Average mileage for two-door and four-door vehicles**

```python
two_door_df = data[data['num-of-doors'] == 2]
four_door_df = data[data['num-of-doors'] == 4]

avg_mileage_two_door = two_door_df[['city-mpg', 'highway-mpg']].mean()
avg_mileage_four_door = four_door_df[['city-mpg', 'highway-mpg']].mean()


categories = ['City MPG', 'Highway MPG']
two_door_values = avg_mileage_two_door.values
four_door_values = avg_mileage_four_door.values

plt.figure(figsize=(10, 6))
plt.bar(categories, two_door_values, width=0.4, label='Two-Door')
plt.bar(categories, four_door_values, width=0.4, label='Four-Door', align='edge')
plt.xlabel('Mileage Type')
plt.ylabel('Average Mileage')
plt.title('Comparison of Average Mileage between Two-Door and Four-Door Vehicles')

plt.legend()
plt.show()
```
